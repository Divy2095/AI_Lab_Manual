```python

# ==============================
# DOWNLOAD REQUIRED NLTK DATA
# ==============================
nltk.download('punkt')
nltk.download('punkt_tab')   # ✅ Tokenizer fix
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')   # ✅ FINAL FIX

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk import pos_tag, RegexpParser
```

```python

# ==============================
# LOAD SPACY MODEL
# ==============================
nlp = spacy.load("en_core_web_sm")
```

```python

# ==============================
# INPUT TEXT
# ==============================
text = input("Enter a sentence: ")
print("\nOriginal Text:\n", text)
```

```python
# ==============================
# 1. TOKENIZATION
# ==============================
tokens = word_tokenize(text)
print("\n1. Original Tokens:\n", tokens)
```

```python

# ==============================
# 2. FILTRATION (REMOVE PUNCTUATION)
# ==============================

filtered_tokens = [word for word in tokens if word not in string.punctuation]
print("\n2. Filtered Tokens:\n", filtered_tokens)
```

```python
# ==============================
# 3. STOPWORDS REMOVAL
# ==============================
stop_words = set(stopwords.words('english'))
clean_tokens = [word for word in filtered_tokens if word.lower() not in stop_words]
print("\n3. After Stopword Removal:\n", clean_tokens)
```

```python

# ==============================
# 4. POS TAGGING
# ==============================
pos_tags = pos_tag(clean_tokens)
print("\n4. POS Tags:\n", pos_tags)
```

```python
# ==============================
# 5. NOUN PHRASE CHUNKING
# ==============================
grammar = "NP: {<DT>?<JJ><NN.>+}"
chunk_parser = RegexpParser(grammar)
chunk_tree = chunk_parser.parse(pos_tags)

print("\n5. Noun Phrase Chunking:\n")
print(chunk_tree)
```

```python

# ==============================
# 6. DEPENDENCY PARSING (spaCy)
# ==============================
doc = nlp(text)

print("\n6. Dependency Parsing:\n")
for token in doc:
    print(f"{token.text} --> {token.dep_} --> {token.head.text}")
```

```python
# Take your own document of 10 sentences and perform the same tasks. Attach code and screenshot of your output.
import nltk
import string
import spacy

# Download required data
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('wordnet')
nltk.download('omw-1.4')

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk import pos_tag, RegexpParser
from nltk.stem import PorterStemmer, WordNetLemmatizer

# Load spacy
nlp = spacy.load("en_core_web_sm")

# INPUT DOCUMENT (10 sentences)
text = """I love eating pizza on weekends.
The weather is very hot today.
She is going to college by bus.
This mobile phone is very expensive.
I completed my project yesterday.
He plays cricket every evening.
The movie was really interesting.
My friend lives in another city.
We are learning programming together.
The food in this restaurant is delicious."""

print("Original Text:\n", text)

# TOKENIZATION
tokens = word_tokenize(text)
print("\nTokens:\n", tokens)

# FILTERING
filtered = [w for w in tokens if w not in string.punctuation]

# STOPWORDS REMOVAL
stop_words = set(stopwords.words('english'))
clean_tokens = [w for w in filtered if w.lower() not in stop_words]
print("\nClean Tokens:\n", clean_tokens)

# POS TAGGING
pos_tags = pos_tag(clean_tokens)
print("\nPOS Tags:\n", pos_tags)

# NOUN PHRASE CHUNKING
grammar = "NP: {<DT>?<JJ>*<NN.*>+}"
cp = RegexpParser(grammar)
tree = cp.parse(pos_tags)
print("\nChunk Tree:\n", tree)

# DEPENDENCY PARSING
doc = nlp(text)
print("\nDependency Parsing:\n")
for token in doc:
    print(token.text, "->", token.dep_, "->", token.head.text)

# STEMMING
stemmer = PorterStemmer()
stemmed = [stemmer.stem(word) for word in clean_tokens]

print("\nStemming:\n", stemmed)

# LEMMATIZATION
lemmatizer = WordNetLemmatizer()
lemmatized = [lemmatizer.lemmatize(word) for word in clean_tokens]

print("\nLemmatization:\n", lemmatized)
```
