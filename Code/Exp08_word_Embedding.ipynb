```python
sentences = [
  "India is a country known for its rich culture and diversity.",
  "The capital of India is New Delhi.",
  "India is famous for festivals, food, and historical places.",

  "Cricket is one of the most popular games in India.",
  "Playing games helps improve teamwork and concentration.",

  "Technology makes our daily life faster and more connected. ",
  "AI subject makes our life smarter."
]
```

```python
#IMPORTING BASIC LIBRARIES

import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer #TF-BoW
from sklearn.feature_extraction.text import TfidfVectorizer #TF-BoW
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#TF-BoW Embedding
tf_vectorizer=CountVectorizer()
tf_emb=tf_vectorizer.fit_transform(sentences)
```

```python
#similarity matrix using cosine_similarity
tf_sim=cosine_similarity(tf_emb)
```

```python
tf_sim
```

```python
#TF-IDF Embedding
tfidf_vectorizer=TfidfVectorizer()
tfidf_emb=tfidf_vectorizer.fit_transform(sentences)
```

```python
tfidf_sim=cosine_similarity(tfidf_emb)
```

```python
tfidf_sim
```

```python
#Recommend the most matching/ similar sentence
def doc_sim(tfidf_sim,sentences):
  for i in range(len(sentences)):
    sim=tfidf_sim[i].copy()
    sim[i]=-1
    top_index=np.argmax(sim)
    print(sentences[i],"->",sentences[top_index])
```

```python
doc_sim(tfidf_sim,sentences)
```
