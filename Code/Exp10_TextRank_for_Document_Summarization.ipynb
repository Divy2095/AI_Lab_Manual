```python
# Install libraries
!pip install pytextrank spacy
!python -m spacy download en_core_web_sm

# Import libraries
import spacy
import pytextrank

# Load model
nlp = spacy.load("en_core_web_sm")
nlp.add_pipe("textrank")

# Input text (portfolio)
text = """
My name is Meet Limbani and I am a B.Tech ICT student at Marwadi University.
I completed my 12th from Gangotri school of Science.
I am interested in cloud computing and want to build my career in this field.
My goal is to get a good job and earn money.
I also want to start my own business in future.
My hobbies are playing cricket and volleyball.
I have played volleyball at district level and became a district champion.
I enjoy playing mobile games and watching movies and web series.
I am helpful and like to support people who are struggling.
Sometimes I feel angry and frustrated but I try to improve myself.
"""

# Apply TextRank
doc = nlp(text)

# Print top sentences (summary)
print("Top Ranked Sentences:\n")
for sent in doc._.textrank.summary(limit_phrases=10, limit_sentences=5):
    print(sent)
```

```python
# Input text (portfolio)
text = """
My name is Meet Limbani, and I am a B.Tech ICT student at Marwadi University. I completed my 12th from Gangotri of Science. I am interested in cloud computing and want to build my career in this field. My goal is to get a good job and earn money. I also want to start my own business in the future. My hobbies include playing cricket and volleyball. I have played volleyball at the district level and became a district champion. I also enjoy playing mobile games and watching movies and web series. I am helpful by nature and like to support people who are struggling. Sometimes I feel angry and frustrated, but I always try to improve myself.
"""

# Apply TextRank
doc = nlp(text)

# Calculate the number of sentences in the original text
original_sentence_count = len(list(doc.sents))
print(f"Original text has {original_sentence_count} sentences.\n")

# Generate 50% summary
target_summary_sentences_50 = max(1, int(original_sentence_count * 0.5))
print(f"Summary (50% of original size, approx {target_summary_sentences_50} sentences):\n")
summary_50_list = []
for sent in doc._.textrank.summary(limit_sentences=target_summary_sentences_50):
    summary_50_list.append(str(sent))
    print(sent)

print("\n" + "-"*50 + "\n")

# Generate 25% summary
target_summary_sentences_25 = max(1, int(original_sentence_count * 0.25))
print(f"Summary (25% of original size, approx {target_summary_sentences_25} sentences):\n")
summary_25_list = []
for sent in doc._.textrank.summary(limit_sentences=target_summary_sentences_25):
    summary_25_list.append(str(sent))
    print(sent)
```

### Comment on the Summary:

The TextRank algorithm focuses on extracting key sentences based on their interconnectedness and importance within the text. It generally does a good job of capturing core information.

For the 50% summary, it highlights your academic background, career interest, a significant achievement (volleyball), and a personal trait (helpfulness). This provides a fairly comprehensive overview.

The 25% summary is more concise, focusing on your identity and a major achievement. This is appropriate for a very brief introduction.

### Which sentence should be there but was not spotted by TextRank?

I believe the sentence: **"My goal is to get a good job and earn money."** or **"I also want to start my own business in the future."** could have been beneficial in the 50% summary. These sentences speak to your future aspirations, which are often crucial in a portfolio. TextRank prioritizes factual statements and achievements, sometimes overlooking forward-looking statements unless they are heavily linked to other core topics.

### Rating the Overall Summary (1-5):

I would rate the overall summary a **4 out of 5**.

The algorithm effectively identifies key sentences relating to identity, education, career interests, and achievements. It provides a coherent and representative summary at both percentage levels.
